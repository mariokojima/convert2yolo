```python
# carrega pacotes
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image


# Função de conversão de coordenatas (pixel) para Yolo Annotation
def convert2yolo(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[1])/2.0
    y = (box[2] + box[3])/2.0
    w = box[1] - box[0]
    h = box[3] - box[2]
    x = x*dw
    w = w*dw
    y = y*dh
    h = h*dh

    return (x,y,w,h)


# baixa a imagem de teste    
!wget https://storage.googleapis.com/ml-planetun/reconhecimento-angulos/car.jpg

# lê a imagem
im = Image.open('car.jpg')

# Crima imagem
fig, ax = plt.subplots()

# Mostra a imagem
ax.imshow(im)

# Shape
width, height = im.size
# print(width)
# print(height)

BOXleft= 670
BOXtop=69   
BOXwidth=3214
BOXheight= 3015


# mostra o bounding box
rect = patches.Rectangle((BOXleft, BOXtop), BOXwidth, BOXheight, linewidth=2, edgecolor='y', facecolor='none')

# adiciona o boundingBox no plot
ax.add_patch(rect)

# plot
plt.show()







box = (BOXleft,BOXleft+BOXwidth,BOXtop,BOXtop + BOXheight)
bb = convert2yolo(im.size, box)
print (bb)
```
